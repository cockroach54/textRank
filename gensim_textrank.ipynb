```python
text.text
```

```python
# parse html
soup = BeautifulSoup(text.text, 'html.parser')
```

```python
print(soup.prettify())
```

```python
article = soup.find(id='articleBodyContents').get_text()
```

```python
pprint(summarize(article, ratio=0.1, split=True))
```

```python
pprint(keywords(article, ratio=0.03))
```
